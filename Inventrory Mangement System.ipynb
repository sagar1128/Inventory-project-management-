# Add New Item into Record

```python
print("*********************************************")
print("          Welcome to Sagar Store           ")
prod_id = str(input("Enter product id:"))
name = str(input("Enter product name:"))
catogary = str(input("Enter product catogary:"))
price = int(input("Enter product price:"))
quantity = int(input("Enter product quantity:"))
discount = int(input("Enter discount on product :"))

item[prod_id] = {'name': name, 'catogary': catogary, 'price': price, 'quantity': quantity,'discount': discount }

# if(record[prod_id]==prod_id):
#     print("product aready")

js = json.dumps(item)

fd = open("data.json",'w')
fd.write(js)
fd.close()
print("Record has been updated")
print(time.ctime())
print("****Thankyou****")
```
